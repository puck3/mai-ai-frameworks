```python
y = df[target]
X = df.drop(columns=[target])

num_cols = ["JoiningYear", "Age", "ExperienceInCurrentDomain"]
cat_cols = ["Education", "City", "PaymentTier", "Gender", "EverBenched"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y
)
```

Функция оценки метрик классификации.

Главной метрикой выбрана ROC AUC, поскольку она оценивает качество ранжирования объектов по вероятности принадлежности к положительному классу. В отличие от метрик, зависящих от фиксированного порога (например, Accuracy или F1), ROC AUC агрегирует результаты по всем возможным порогам и показывает, насколько стабильно модель отделяет положительные объекты от отрицательных. 

```python
def evaluate_classification(y_true, y_pred, y_proba):
    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred),
        "recall": recall_score(y_true, y_pred),
        "f1": f1_score(y_true, y_pred),
        "roc_auc": roc_auc_score(y_true, y_proba)
    }
```

В качестве бейзлайна используется пайплайн, в котором к категориальным признакам применяется OrdinalEncoder (значения кодируются целыми неотрицательными числами), после чего применяется LogisticRegression.

```python
encoder = ColumnTransformer([
    ("cat", OrdinalEncoder(), cat_cols),
    ("num", "passthrough", num_cols)
])

base = Pipeline([
    ("preprocess", encoder),
    ("lr", LogisticRegression(max_iter=1000))
])

base.fit(X_train, y_train)

y_pred = base.predict(X_test)
y_proba = base.predict_proba(X_test)[:, 1]

baseline_metrics = evaluate_classification(y_test, y_pred, y_proba)
baseline_metrics
```

### Имплементация логистической регрессии

```python
class CustomLogisticRegression(BaseEstimator, ClassifierMixin):
    def __init__(self, C=1.0, l1_ratio=0.5, max_iter=1000, tol=1e-4, lr=0.1):
        self.C = C
        self.l1_ratio = l1_ratio
        self.max_iter = max_iter
        self.tol = tol
        self.lr = lr

    def _sigmoid(self, z):
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))

    def _compute_gradient(self, X, y, y_pred, coef):
        n_samples = X.shape[0]
        error = y_pred - y
        grad = X.T @ error / n_samples
        l1_grad = np.sign(coef) * self.l1_ratio
        l2_grad = coef * (1 - self.l1_ratio)
        grad += (l1_grad + l2_grad) / (self.C * n_samples)
        return grad

    def fit(self, X, y):
        self.classes_ = np.unique(y)
        n_samples, n_features = X.shape
        y_binary = (y == self.classes_[1]).astype(float)
        X = np.hstack([np.ones((n_samples, 1)), X])
        n_features += 1
        self.coef_ = np.zeros(n_features)

        for _ in range(self.max_iter):
            z = X @ self.coef_
            y_pred = self._sigmoid(z)
            grad = self._compute_gradient(X, y_binary, y_pred, self.coef_)
            coef_new = self.coef_ - self.lr * grad
            if np.linalg.norm(coef_new - self.coef_) < self.tol:
                break
            self.coef_ = coef_new

        self.intercept_ = self.coef_[0]
        self.coef_ = self.coef_[1:]
        self.is_fitted_ = True
        return self

    def predict_proba(self, X):
        z = X @ self.coef_ + self.intercept_
        proba_pos = self._sigmoid(z)
        return np.column_stack([1 - proba_pos, proba_pos])

    def predict(self, X):
        proba = self.predict_proba(X)
        y_pred = (proba[:, 1] >= 0.5).astype(int)
        return self.classes_[y_pred]
```

```python
encoder = ColumnTransformer([
    ("cat", OrdinalEncoder(), cat_cols),
    ("num", "passthrough", num_cols)
])

custom_base = Pipeline([
    ("preprocess", encoder),
    ("lr", CustomLogisticRegression())
])

custom_base.fit(X_train, y_train)

y_pred = custom_base.predict(X_test)
y_proba = custom_base.predict_proba(X_test)[:, 1]

custom_base_metrics = evaluate_classification(y_test, y_pred, y_proba)
custom_base_metrics
```

### Имплементация ElasticNet

```python
class CustomElasticNet(BaseEstimator, RegressorMixin):
    def __init__(self, alpha=1.0, l1_ratio=0.5, max_iter=1000, random_state=None):
        self.alpha = alpha
        self.l1_ratio = l1_ratio
        self.max_iter = max_iter
        self.random_state = random_state

    def _soft_threshold(self, x, threshold):
        return np.sign(x) * np.maximum(np.abs(x) - threshold, 0)

    def _compute_regularization(self):
        alpha_l1 = self.alpha * self.l1_ratio
        alpha_l2 = self.alpha * (1 - self.l1_ratio)
        return alpha_l1, alpha_l2

    def fit(self, X, y):
        X = np.array(X, dtype=float)
        y = np.array(y, dtype=float)
        n_samples, n_features = X.shape

        self.X_mean_ = X.mean(axis=0)
        self.y_mean_ = y.mean()
        X_centered = X - self.X_mean_
        y_centered = y - self.y_mean_

        rng = np.random.RandomState(self.random_state)
        coef = rng.randn(n_features) * 0.01

        alpha_l1, alpha_l2 = self._compute_regularization()
        X_squared = np.sum(X_centered**2, axis=0)
        X_squared[X_squared == 0] = 1

        for iteration in range(self.max_iter):
            coef_old = coef.copy()
            for j in range(n_features):
                X_j = X_centered[:, j]
                y_pred = X_centered.dot(coef)
                r_j = y_centered - y_pred + X_j * coef[j]
                z_j = np.dot(X_j, r_j) / n_samples
                if alpha_l2 > 0:
                    coef[j] = self._soft_threshold(z_j, alpha_l1) / (X_squared[j] / n_samples + alpha_l2)
                else:
                    coef[j] = self._soft_threshold(z_j, alpha_l1) * n_samples / X_squared[j]
            if np.max(np.abs(coef - coef_old)) < 1e-4:
                break

        self.coef_ = coef
        self.n_iter_ = iteration + 1
        self.intercept_ = self.y_mean_ - np.dot(self.X_mean_, self.coef_)
        return self

    def predict(self, X):
        X = np.array(X, dtype=float)
        return X.dot(self.coef_) + self.intercept_
```

Обучение собственной модели на неподготовленных данных
